```python
import heapq

INF = int(1e9)
n,m = map(int, input().split())
start = int(input())
graph = [[] for i in range(n+1)]
visited = [False]*(n+1)
distance = [INF]*(n+1)

for _ in range(m):
    a,b,c = map(int, input().split())
    #a번 노드에서 b를 가는데 c 만큼 cost
    graph[a].append((b,c))

def dijkstra(start):
    q=[]
    heapq.heappush(q, ((0, start))
    distance[start] = 0
               
    while q:
        dist, now = heapq.heappop(q):
        #이미 처리된 적이 있는지?
        if distance[now] < dist:
            continue
        for i in graph[now]:
            cost = dist+i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))
               
```
