```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
import sys
input = sys.stdin.readline

n = int(input())
rg,cg = map(int, input().split())
rp,cp = map(int, input().split())
rg,cg = rg-1, cg-1
rp,cp = rp-1, cp-1
gamejam=[]
for i in range(n):
	gamejam.append(list(input().split()))

visited=[[0 for _ in range(n)] for _ in range(n)]

def gamejam_start(nx,ny):
	loc = gamejam[nx][ny]
	visited[nx][ny] = 1
	m,d = int(loc[:-1]), loc[-1:]
	#print('!',m,d)
	if d =="D":
		return m,1,0
	if d=="U":
		return m,-1,0
	if d=="L":
		return m,0,-1
	if d=="R":
		return m,0,1
	
# 시작점에서 방향을 기준으로 움직임
def gamejam_play(move,x,dx,y,dy):
	# 움직임 시작 - 할당받은 수 만큼
	for i in range(move):
		x += dx
		y += dy
		if x<0:
			x = n-1
		elif x>=n:
			x = 0
		if y < 0:
			y = n-1
		elif y>=n:
			y = 0
		#print('#',x,y, dx,dy)
		#방문한 곳을 재방문하게 되면 게임 종료 신호 보냄
		if visited[x][y]:
			return -1,-1
		visited[x][y] = 1
	# 역할을 모두 수행하고, 계속 게임은 진행되고 있음 
	# 마지막 탐색 위치 전송
	return x,y
def total_score(visited):
	temp = 0
	for i in range(n):
		temp += sum(visited[i])
	#print("total:",temp)
	return temp

goorm = 0
player = 0
while True:
	m,dx,dy = gamejam_start(rg,cg)
	rg,cg = gamejam_play(m,rg,dx,cg,dy)
	#print("@",rg,cg)
	if rg ==-1:
		goorm = total_score(visited)
		break
#print(visited)
visited=[[0 for _ in range(n)] for _ in range(n)]
#print(visited)
while True:
	m,dx,dy = gamejam_start(rp,cp)
	rp,cp = gamejam_play(m,rp,dx,cp,dy)
	#print("@",rp,cp)
	if rp ==-1:
		player = total_score(visited)
		break
#print(visited)
if goorm > player:
	print("goorm", goorm)
else:
	print("player", player)

#print(goorm, player)


	
```
